```python
# Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\sidra\OneDrive\Documents\Projects + get them ready fot github\movies.csv')
```

```python
df.head()
```

```python
# Data Cleaning
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
print(df['budget'].skew())
```

```python
df['budget'] = df['budget'].fillna(df['budget'].median())
df['gross'] = df['gross'].fillna(df['gross'].median())
df['runtime'] = df['runtime'].fillna(df['runtime'].median())
df['votes'] = df['votes'].fillna(df['votes'].median())
df['country'] = df['country'].fillna('unknown')
```

```python
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
df.dtypes
```

```python
# Data Types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
import re

# When writing what i learned about my project write...
# df['Country'] = df['released'].str.extract(r'\((.*?)\)')
# df['released'] = df['released'].str.replace(r'\s*\(.*?\)', '', regex=True).copy
# I am modifying the original released column and since the new column is extracted from that, pandas treats them as referencing the same
# data
df['Country'] = df['released'].str.extract(r'\((.*?)\)').copy()
```

```python
df['released'] = df['released'].str.replace(r'\s*\(.*?\)', '', regex=True)
```

```python
df.drop(columns=['country'], inplace=True)
```

```python
# Reorder Columnscols 
cols = list(df.columns)
cols.insert(cols.index('released') + 1, cols.pop(cols.index('Country')))
df = df[cols]
```

```python
df['released'] = pd.to_datetime(df['released'],
                                errors = 'coerce')
```

```python
df
```

```python
df['correct_year'] = df['released'].astype(str).str[0:4]

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Dropping duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
df.drop_duplicates()
```

```python
# Scatter Plot with Budget vs Gross Revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Films')

plt.show()
```

```python
# Scatter Plot with Budget vs Gross Revenue using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
plt.show()
```
